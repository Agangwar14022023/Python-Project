```python
class Hospital_Management:    
    Sp1={}
    Sp2={}
    Sp3={}
    Sp4={}
    Sp5={}
    def __init__(self):
        print("""===========================================================================================================================
                                     ⫸ ➕➕➕ WELCOME TO THE CITY HOSPITAL ➕➕➕ ⫷               
============================================================================================================================""")
        print('''List for Managing Record of Patients
============================================================================================================================''')
#         self.Sp1={}
#         self.Sp2={}
#         self.Sp3={}
#         self.Sp4={}
#         self.Sp5={}
        self.L=int(input('1. Add New Patient \n2. Print all Patients \n3. Get next Patients \n4. Remove a leaving Patient \n5. End The Program \nEnter Your Choice From (1 to 5): '))
        if self.L==1:
            self.Add_New_Patient()
        elif self.L==2:
            self.Print_All_Patients()
        elif self.L==3:
            self.Get_Next_Patient()
        elif self.L==4:
            self.Remove_Leaving_Patients()
        elif self.L==5:
            self.End_The_Program()
            
        
    
    def Add_New_Patient(self):
        global Pt
        global Sp1
        global Sp2
        global Sp3
        global Sp4
        global Sp5
        print('============================================================================================================================')
        Sp = int(input('Enter Your Specialization From (1 to 5): '))
        Pt = input('Enter Patient Name: ')
        Ag = int(input('Enter Patient Age: '))
        St = int(input('Enter Status \n 1. Normal \n 2. Urgent \n 3. Very Urgent \n Choose Your Status Number From Above: '))
        print()
    #Specialization
        if St==1:
            St= 'Normal Status'
        elif St==2:
            St= 'Urgent Status'
        elif St==3:
            St= 'Very Urgent Status'
        if Sp==1:
            self.Sp1[Pt]=St
            l=sorted(self.Sp1.items(),key=lambda x:x[1],reverse=True)
            self.Sp1=dict(l)           
        elif Sp==2:
            self.Sp2[Pt]=St
            l=sorted(self.Sp2.items(),key=lambda x:x[1],reverse=True)
            self.Sp2=dict(l)
        elif Sp==3:
            self.Sp3[Pt]=St
            l=sorted(self.Sp3.items(),key=lambda x:x[1],reverse=True)
            self.Sp3=dict(l)
        elif Sp==4:
            self.Sp4[Pt]=St
            l=sorted(self.Sp4.items(),key=lambda x:x[1],reverse=True)
            self.Sp4=dict(l)
        elif Sp==5:
            self.Sp5[Pt]=St
            l=sorted(self.Sp5.items(),key=lambda x:x[1],reverse=True)
            self.Sp5=dict(l)
            
        print('Patient Added in Database Sucessfully')
        print('============================================================================================================================')
        
        C = input('Do you wish to go back to main menu ? (y/n): ')
        if C=='y':
            self.__init__()
        else:
            return
    def Print_All_Patients(self):
        global Sp1
        global Sp2
        global Sp3
        global Sp4
        global Sp5
        print('============================================================================================================================')
        print('Specialization 1: There are {} Patients.'.format(len(self.Sp1)))
        print(self.Sp1)
        print()
        print('Specialization 2: There are {} Patients.'.format(len(self.Sp2)))
        print(self.Sp2)
        print()
        print('Specialization 3: There are {} Patients.'.format(len(self.Sp3)))
        print(self.Sp3)
        print()
        print('Specialization 4: There are {} Patients.'.format(len(self.Sp4)))
        print(self.Sp4)
        print()
        print('Specialization 5: There are {} Patients.'.format(len(self.Sp5)))
        print(self.Sp5)
        print('============================================================================================================================')
        C = input('Do you wish to go back to main menu ? (y/n): ')
        if C=='y':
            self.__init__()
        else:
            return
        
    def Get_Next_Patient(self):
        global Pt
        global Sp1
        global Sp2
        global Sp3
        global Sp4
        global Sp5
        print('============================================================================================================================')
        Sp = int(input('Enter Your Specialization From (1 to 5): '))
        if Sp==1:
#            l=list(self.Sp1.items())
#            l.sort() 
#            self.Sp1=dict(l)
            if len(self.Sp1) !=0:              
                print(self.Sp1[Pt][::],'===> Please Go With Dr')
            else:
                print('No Patients at the Moment. Have rest, Dr')
        elif Sp==2:
#            l=list(self.Sp2.items())
#            l.sort() 
#            self.Sp2=dict(l)
            if len(self.Sp2) !=0:              
                print(self.Sp2[Pt][::],'===> Please Go With Dr')
            else:
                print('No Patients at the Moment. Have rest, Dr')
        elif Sp==3:
#            l=list(self.Sp3.items())
#            l.sort() 
#            self.Sp3=dict(l)
            if len(self.Sp3) !=0:              
                print(self.Sp3[Pt][::],'===> Please Go With Dr')
            else:
                print('No Patients at the Moment. Have rest, Dr')
        elif Sp==4:
#            l=list(self.Sp4.items())
#            l.sort() 
#            self.Sp4=dict(l)
            if len(self.Sp4) !=0:              
                print(self.Sp4[Pt][::],'===> Please Go With Dr')
            else:
                print('No Patients at the Moment. Have rest, Dr')
        elif Sp==5:
#            l=list(self.Sp5.items())
#            l.sort() 
#            self.Sp5=dict(l)
            if len(self.Sp5) !=0:              
                print(self.Sp5[Pt][::],'===> Please Go With Dr')
            else:
                print('No Patients at the Moment. Have rest, Dr')
            
        print('============================================================================================================================')
        C = input('Do you wish to go back to main menu ? (y/n): ')
        if C=='y':
            self.__init__()
        else:
            return
    
    def Remove_Leaving_Patients(self):
        print('============================================================================================================================')
        self.Sp = int(input('Enter Your Specialization From (1 to 5): '))
        self.Pt = input('Enter Patient Name: ')
        if self.Sp==1:
            if self.Pt in self.Sp1:
                del self.Sp1[Pt]
                print('Successfully Removed')
            else:
                print('No Patient with such a name in this Specialization')
        elif self.Sp==2:
            if self.Pt in self.Sp2:
                del self.Sp2[Pt]
                print('Successfully Removed')
            else:
                print('No Patient with such a name in this Specialization')
        elif self.Sp==3:
            if self.Pt in self.Sp3:
                del self.Sp3[Pt]
                print('Successfully Removed')
            else:
                print('No Patient with such a name in this Specialization')
        elif self.Sp==4:
            if self.Pt in self.Sp4:
                del self.Sp4[Pt]
                print('Successfully Removed')
            else:
                print('No Patient with such a name in this Specialization')
        elif self.Sp==5:
            if self.Pt in self.Sp5:
                del self.Sp5[Pt]
                print('Successfully Removed')
            else:
                print('No Patient with such a name in this Specialization')
                
        print('============================================================================================================================')
        C = input('Do you wish to go back to main menu ? (y/n): ')
        if C=='y':
            self.__init__()
        else:
            return
        
    def End_The_Program(self):
        return
        
       
        
                   
a=Hospital_Management()
print(a.Sp3)
```
